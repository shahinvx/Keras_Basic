```
# Equation of line / Hypothesis
def line_b(x, w, b):
  return x * w + b
```

```
# Cost function
def mean_square_error(y, y_change):
  sqr = (y - y_change)**2
  return sqr.mean()
```

```
x = df[['Height']].values
y = df[['Weight']].values
y
```

```
y_change = line_b(x, w=2, b=2)
mean_square_error(y, y_change)
```

```
# change b to get the close result or beter model

x = df[['Height']].values
y = df[['Weight']].values

plt.figure(figsize=(11,5))

ax1 = plt.subplot(121)
df.plot(kind='scatter',
        x = 'Height',
        y = 'Weight',
        title='Weight and Height in Age', ax = ax1)

b_change = np.array([-100,-50, -30, -20, 0, 50, 100, 150])
m_error = []
for b in b_change:
  y_change = line_b(x, w=2, b=b)
  err = mean_square_error(y, y_change)
  m_error.append(err)
  plt.plot(x,y_change)
print(m_error)
ax2 = plt.subplot(122)
plt.plot(b_change, m_error, 'o-')
plt.title('Cost as function of b')
plt.xlabel('b')
```

```
# change b and w to get the close result or beter model

x = df[['Height']].values
y = df[['Weight']].values

plt.figure(figsize=(11,5))

ax1 = plt.subplot(121)
df.plot(kind='scatter',
        x = 'Height',
        y = 'Weight',
        title='Weight and Height in Age', ax = ax1)

b_change = np.array([-100,-50, -30, -20, 0, 50, 100, 150])
w_change = np.array([.1,.2,.3,.4,.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5,5, 5.5,6,6.5 ,7,7.5,8,8.5,9,9.5,10,10.5,11,11.5,12,12.5,13,13.5])
m_error = []
for b in b_change:
  mn = 999999999999999999999999999999999
  for w in w_change:
    y_change = line_b(x, w=w, b=b)
    err = mean_square_error(y, y_change)
    mn = min(mn,err)
    plt.plot(x,y_change)
  m_error.append(mn)

print(m_error)
ax2 = plt.subplot(122)
plt.plot(b_change, m_error, 'o-')
plt.title('Cost as function of b and w')
plt.xlabel('b')
```

<h2> Linear Regrassion with Keras


```
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, SGD
```

```
df = pd.read_csv('/content/drive/My Drive/Google Colab/Neural Network/Keras/My_Code_Keras/weight-height.csv')
x = df[['Height']].values
y = df[['Weight']].values
```

```
model = Sequential()
model.add(Dense(1, input_shape=(1,))) # Layer # Dense = 1 means y_change as out
model.summary()
```

```
model.compile(Adam(lr=.8), 'mean_squared_error') # Optimizer , learning rate, loss function
```

```
model.fit(x, y, epochs=60)
```

```
y_change = model.predict(x) # get the trained value of Y
```

```
df.plot(kind='scatter',
        x = 'Height',
        y = 'Weight',
        title='Weight and Height in Age')

plt.plot(x , y_change, color='red',linewidth=3)
```

```
w , b = model.get_weights()  # We can get the trained value of weight and bias
print("Adjusted weight ",w)
print("Adjusted bias ",b)
```

```
y_change = line(x, w, b)
df.plot(kind='scatter',
        x = 'Height',
        y = 'Weight',
        title='Weight and Height in Age')
plt.plot(x , y_change, color='red',linewidth=3)
```
