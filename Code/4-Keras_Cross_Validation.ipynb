<h2> <center><font color="#0099cc"> Cross Validation (CV)<br></center>
<h5>* Take all possible test and train split from our data set and make the best score using KFold method.</h5>
<h5>* Each Stratified KFold has same ratio of True and False<h5>


<h5>-> (LOLO) Leave One Level Out (CV)<h5>
<h5>-> (LPLO) Leave P Level's Out (CV)<h5>

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, SGD
```

```
df = pd.read_csv('/content/drive/My Drive/Google Colab/Neural Network/Keras/My_Code_Keras/user_visit_duration.csv')
df.head()
```

```
from keras.wrappers.scikit_learn import KerasClassifier
```

```
def build_logistic_regrassion_model():
  model = Sequential()
  model.add(Dense(1, input_shape=(1,), activation='sigmoid'))
  model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics=['accuracy'])
  return model
```

```
model = KerasClassifier(build_fn=build_logistic_regrassion_model, epochs=25, verbose=0)
```

```
from sklearn.model_selection import cross_val_score, KFold
```

```
x = df[['Time (min)']].values
y = df[['Buy']].values
```

```
cv = KFold(n_splits=3, shuffle=True)
score = cross_val_score(model, x, y, cv = cv)
score
```

```
print("Cross validation accuracy is : {:0.4f}".format(score.mean()),"± {:0.4f}".format(score.std()))
```
