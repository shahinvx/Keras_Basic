<h2><center><font color='Aquamarine'>Confusion Matrics</h2></font><center><h4>
We can going to Beyond accuracy score and introduce better way to judge , how well our model is !!

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, SGD
```

```
df = pd.read_csv('/content/drive/My Drive/Google Colab/Neural Network/Keras/My_Code_Keras/user_visit_duration.csv')
df.head()
```

```
model = Sequential()
model.add(Dense(1, input_shape=(1,), activation='sigmoid'))
model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```
x = df[['Time (min)']].values
y = df['Buy'].values

model.fit(x, y, epochs=25)
```

<h2><center>Confusion Matrics
<h5>* Over All Accuracy : (Total Positive + Total Negetive) / Total # (TP + TN) / total
<h5>* Precision : TP / test yes
<h5>* Recall : TP / actual yes
<h5>* F1 : 2 * (P*R) / (P+R)

```
y_pred = model.predict(x)
y_class_pred = y_pred> 0.5
```

```
from sklearn.metrics import confusion_matrix
```

```
confusion_matrix(y, y_class_pred)
```

```
def pretty_confusion_matrix(y, y_pred, labels=["False", "True"]):
  cm = confusion_matrix(y, y_pred)
  pred_labels = ['Predicted '+ l for l in labels]
  df = pd.DataFrame(cm, index=labels, columns=pred_labels)
  return df
```

```
pretty_confusion_matrix(y, y_class_pred, ['Not Buy' , 'Buy'])
```

```
from sklearn.metrics import precision_score, recall_score, f1_score
```

```
print("Precision score :\t{:0.3f}".format(precision_score(y, y_class_pred))," 43 / (43+11)")
print("Recall score :\t\t{:0.3f}".format(recall_score(y, y_class_pred)), " 43 / (43+7)")
print("F1 score :\t\t{:0.3f}".format(f1_score(y, y_class_pred)), " 2*(P*R) / (P+R)")
```

```
from sklearn.metrics import classification_report
```

```
print(classification_report(y, y_class_pred))
```
